```python
import pandas as pd
import regex as re

import spacy
nlp = spacy.load("en_core_web_md")
```

```python
df = pd.read_csv("output_claude.csv")
```

```python
# Splitting columns and rows
rows = [str.split('\n') for str in df.name]
result = []
for row in rows:
    columns = [re.split(',(?![ \t])', r) for r in row]
    result.extend(columns)

result
```

```python
# Creating a dataframe with named columns
df = pd.DataFrame(result)
df.rename(columns={0: 'name', 1: 'country', 2: 'age', 3: 'occupation', 4: 'description', \
                   5: 'shadow'}, inplace=True)
df.head()
```

```python
# Adjust age of death by estimated climate change impacts

df_age = pd.read_csv('age_corrected.csv')
df_age['integer'] = df['age'].astype(int)
df_age = df_age.drop('country', axis=1)
df_age = df_age.drop('age', axis=1)

df_new = df.merge(df_age, how='left', on='name')
df_new = df_new.drop('integer', axis=1)
df_new
```

```python
df_new.to_csv('output_tidy.csv', index=False)
```

```python
df = pd.read_csv('output_tidy.csv')
df_scores = pd.read_csv('output_scores.csv')
df = df.merge(df_scores, how='left', on='name')

df
```

```python
df.describe()
```

```python
cliques = {'perfectionists': ["Bai Yang", "Qiu Yue", "Ren Nakamura", "Felix Becker", "Olivia Patel"], \
           'family': ["Xue Fang", "Jian Wen", "Mei Lin", "Shen Tao", "Divya Mehta", "Putri Dewi"], \
            'artists': ["Anika Gupta", "Imran Bhatti", "Sofía Torres", "Luca Esposito"], \
            'progressives': ["Nikhil Chopra", "Chimamanda Obi", "Gustavo Santos", "Ava Rodriguez"], \
            'resilient': ["Saanvi Kulkarni", "Nusrat Jahan", "Ebele Onwudiwe", "Wayan Kartika"]}
```

```python
# Adding a column to describe their clique
df['clique'] = df['name'].map({v: k for k, lst in cliques.items() for v in lst})
df.fillna('loner', inplace=True)
df
```

```python
# Trait means of different cliques
open_mean = df.groupby('clique').Openness.mean()
consc_mean = df.groupby('clique').Conscientiousness.mean()
extra_mean = df.groupby('clique').Extraversion.mean()
agre_mean = df.groupby('clique').Agreeableness.mean()
neuro_mean = df.groupby('clique').Neuroticism.mean()
pop_mean = df.groupby('clique').Popularity.mean()

print(open_mean, consc_mean, extra_mean, agre_mean, neuro_mean, pop_mean)
```
